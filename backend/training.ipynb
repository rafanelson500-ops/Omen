```python
import numpy as np
from classes.datastream import Datastream
from classes.features import add_context_features
from hmmlearn.hmm import GaussianHMM
import joblib
```

```python
def get_data(date):
    datastream = Datastream()
    _, _, long_df = datastream.historical(date + "T13:25:00.000000000Z", date + "T14:00:00.000000000Z")
    return long_df
```

```python
features = ["efficiency_ratio", "vol_ratio", "autocorr_1", "vwap_z"]

dates = ["2026-4-1", "2026-3-31", "2026-3-30"]

X_list = []
lengths = []
for date in dates:
    sample = add_context_features(get_data(date))
    X_list.append(sample[features].values)
    lengths.append(len(sample))

X = np.vstack(X_list)

model = GaussianHMM(n_components=2, covariance_type="full", n_iter=100, random_state=42)
model.fit(X, lengths)

joblib.dump(model, "models/hmm.pkl")
```
